## Copy model

## Connecting populations

https://nest-simulator.readthedocs.io/en/v3.4/synapses/connection_management.html#connection-management

```python
d = 0.1
conn_dict_ex = {"rule": "fixed_indegree", "indegree": 20}
syn_dict_ex = {"delay": d, "weight": 2.0} #static synapse "synapse_model":"static_synapse"
nest.Connect(epop1, ipop1, conn_dict_ex, syn_dict_ex)
```

```python
conn_dict_in = {"rule": "fixed_indegree", "indegree": 12}
syn_dict_in = {"delay": d, "weight": -4.0} #static synapse
nest.Connect(ipop1, epop1, conn_dict_in, syn_dict_in)
```

```python
p_rate = 15000.0
noise = nest.Create("poisson_generator", params={"rate": p_rate})
```

```python
nest.Connect(noise, epop1, syn_spec=syn_dict_ex) #by default conn_spec is all-to-all rule.
nest.Connect(noise, ipop1, syn_spec=syn_dict_ex)
```

## Adding spatial information

https://nest-simulator.readthedocs.io/en/v3.4/networks/spatially_structured_networks.html#spatial-networks

```python
positions = nest.spatial.grid(shape=[10, 10],  # the number of rows and column in this grid ...
                              extent=[2., 2.]  # the size of the grid in mm
                              )
```

```python
s_nodes1 = nest.Create('iaf_psc_alpha', positions=positions)
```

```python
positions = nest.spatial.free(
    nest.random.uniform(min=-0.5, max=0.5),  # using random positions in a uniform distribution
    num_dimensions=2 #3  # have to specify number of dimensions
)
```

```python
s_nodes2 = nest.Create('iaf_psc_alpha', 800, positions=positions)
```

```python
nest.PlotLayer(s_nodes1)
plt.show()
```

```python
nest.PlotLayer(s_nodes2)
plt.show()
```

```python
s_nodes2_inh = nest.Create('iaf_psc_alpha', 200, positions=positions)
```

```python
nest.PlotLayer(s_nodes2_inh)
plt.show()
```

```python
conn_dict_ex = {'rule': 'pairwise_bernoulli',
                'p': 1.0,
                'mask': {'circular': {'radius': 0.4}}}
nest.Connect(s_nodes2, s_nodes2_inh, conn_dict_ex) # And now we connect E->I, weight by default is 1
```

```python
conn_dict_in = {'rule': 'pairwise_bernoulli',
                'p': 1.0,
                'mask': {'circular': {'radius': 0.65}}
               }

syn_spec={'weight': -1.0*nest.random.uniform(0.5, 2.)}

# connecting the rest of the populations:
nest.Connect(s_nodes2, s_nodes2, conn_dict_ex)
nest.Connect(s_nodes2_inh, s_nodes2_inh, conn_dict_in, syn_spec)
nest.Connect(s_nodes2_inh, s_nodes2, conn_dict_in, syn_spec)
```

## Collocated synapses

https://nest-simulator.readthedocs.io/en/v3.4/synapses/connection_management.html#collocated-synapses

```python
nest.ResetKernel()
```

```python
nodes1 = nest.Create('iaf_psc_alpha', 3)
nodes2 = nest.Create('iaf_psc_alpha', 3)
```

```python
syn_spec = nest.CollocatedSynapses({'weight': 4.0, 'delay': 1.0}, #static synapse
                                   {'synapse_model': 'stdp_synapse'},
                                   {'synapse_model': 'stdp_synapse', 'tau_plus': 30.0})
```

```python
nest.Connect(nodes1, nodes2, conn_spec='one_to_one', syn_spec=syn_spec)
```

```python
print(nest.GetConnections())
```

```python
print(nest.GetConnections().tau_plus)
```

```python
print(nest.num_connections)
```

```python
nest.ResetKernel()
```

```python
A = nest.Create('iaf_psc_exp_multisynapse', 7)
B = nest.Create('iaf_psc_exp_multisynapse', 7, {'tau_syn': [0.1, 1.1, 2.1, 3.1, 4.1, 5.1, 6.1]}) #[0.1 + i for i in range(7)]})
```

```python
syn_spec_dict = nest.CollocatedSynapses({'weight': 5.0, 'receptor_type': 2},
                                        {'weight': 1.5, 'receptor_type': 7})
```

```python
nest.Connect(A, B, 'one_to_one', syn_spec_dict)
```
